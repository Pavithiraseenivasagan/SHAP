# SHAP Explainability Tutorial
**Dataset:** Breast Cancer Wisconsin Dataset

## Step 1: Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
import shap

sns.set(style='whitegrid')
shap.initjs()
```

## Step 2: Load the Breast Cancer Dataset

```python
data = load_breast_cancer()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target, name='target')
X['target'] = y
X.shape
```

## Step 3: Exploratory Data Analysis (EDA)

### 3.1 Overview of Data

```python
X.head()
```

### 3.2 Summary Statistics

```python
X.describe()
```

### 3.3 Check for Missing Values

```python
X.isnull().sum()
```

### 3.4 Class Distribution

```python
sns.countplot(x='target', data=X, palette='Set2')
plt.title("Target Class Distribution")
plt.xlabel("Class (0 = malignant, 1 = benign)")
plt.ylabel("Count")
plt.show()
```

### 3.5 Correlation Heatmap

```python
plt.figure(figsize=(12, 10))
correlation = X.drop(columns='target').corr()
sns.heatmap(correlation, cmap='coolwarm', linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

## Step 4: Prepare Data for Modeling

```python
X = X.drop(columns='target')
y = data.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Step 5: Train XGBoost Classifier

```python
model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
model.fit(X_train_scaled, y_train)
```

## Step 6: Initialize SHAP Explainer and Compute Values

```python
explainer = shap.Explainer(model, X_train_scaled)
shap_values = explainer(X_test_scaled)
```

## Step 7: SHAP Summary Plot

```python
shap.summary_plot(shap_values, X_test, plot_type='bar')
```

## Step 8: SHAP Beeswarm Plot

```python
shap.summary_plot(shap_values, X_test)
```

## Step 9: SHAP Force Plot for a Single Prediction

```python
shap.initjs()
shap_values_array = shap_values.values
shap.force_plot(explainer.expected_value, shap_values_array[0], X_test.iloc[0], matplotlib=True)
```
